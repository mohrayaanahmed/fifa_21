```python
# import the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import seaborn as sns
pd.options.mode.chained_assignment = None
```

```python
# read the dataset into df
df = pd.read_csv('fifa.csv')
```

```python
# clean the data by deleting the columns not required for the project
del df["player_url"]
del df["sofifa_id"]
del df["dob"]
del df["height_cm"]
del df["weight_kg"]
del df["international_reputation"]
del df["body_type"]
del df["real_face"]
del df["team_jersey_number"]
del df["loaned_from"]
del df["joined"]
del df["contract_valid_until"]
del df["nation_jersey_number"]
del df["defending_marking"]
del df["ls"]
del df["rs"]
del df["lw"]
del df["lf"]
del df["st"]
del df["cf"]
del df["rf"]
del df["rw"]
del df["lam"]
del df["cam"]
del df["ram"]
del df["lm"]
del df["lcm"]
del df["cm"]
del df["rcm"]
del df["rm"]
del df["lwb"]
del df["ldm"]
del df["cdm"]
del df["rdm"]
del df["rwb"]
del df["lb"]
del df["lcb"]
del df["cb"]
del df["rcb"]
del df["long_name"]
del df["league_rank"]
del df["preferred_foot"]
del df["weak_foot"]
del df["skill_moves"]
del df["player_tags"]
del df["player_traits"]
del df["rb"]
del df["team_position"]
del df["nation_position"]
del df["league_name"]
del df["goalkeeping_handling"]
del df["goalkeeping_reflexes"]
del df["goalkeeping_positioning"]
del df["goalkeeping_diving"]
del df["goalkeeping_kicking"]

# print first five records of the dataset
df.head()
```

```python
# 3.2 CENTRE MIDFIELDERS

df_cm = df.loc[df['player_positions'].str.contains("CM", case = True)]
df_cm = df_cm.reset_index(drop = True)

df_cm
```

```python
for i in range(0, len(df_cm['overall'])):
    if df_cm['overall'][i] > 85:
        df_cm = df_cm.drop(i)

df_cm = df_cm.reset_index(drop=True)

# remove irrelevant columns
df_cm = df_cm.drop(['defending_sliding_tackle', 'mentality_positioning', 'nationality', 'shooting', 'defending', 'attacking_heading_accuracy', 'power_strength', 'skill_fk_accuracy', 'attacking_finishing', 'club_name', 'age', 'value_eur', 'wage_eur', 'player_positions', 'release_clause_eur', 'work_rate', 'gk_reflexes', 'gk_positioning', 'gk_handling', 'gk_diving', 'gk_kicking', 'gk_speed'], axis = 1)

# for safekeeping keep the values in cm dataset
cm = df_cm

# get the top 10 rated CM according to their potential
df_cm = df_cm[0:10]
df_cm
```

```python
# plot the top 10 cms with their most associated attributes

df_cm.plot(x = 'short_name', y = ['pace', 'passing', 'dribbling', 'physic', 'attacking_crossing', 'attacking_short_passing', 'attacking_volleys', 'skill_dribbling', 'skill_curve', 'skill_long_passing', 'skill_ball_control', 'movement_acceleration', 'movement_sprint_speed', 'movement_agility', 'movement_reactions', 'movement_balance', 'power_shot_power', 'power_jumping', 'power_stamina', 'power_long_shots', 'mentality_aggression', 'mentality_interceptions', 'mentality_vision', 'mentality_penalties', 'mentality_composure', 'defending_standing_tackle'])
plt.xlabel('Central Midfielder Names', fontsize = 20)
plt.ylabel('Attribute Score', fontsize = 20)
plt.title("The difference of Top 10 Central Midfielders with respect to the attributes imperative for their position")
plt.legend(bbox_to_anchor=(1, 1), fontsize = 7)
plt.show()
```

```python
# get the cm names in names
names_cm = []
names_cm = pd.DataFrame(cm['short_name'])
cm = cm.drop('short_name', axis = 1)
cm = cm.drop('potential', axis = 1)
names_cm.append(df_cm['short_name'])

cm_norm = (cm - cm.mean())/cm.std()
```

```python
# use Linear Regression to predict the overall of CMs
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
cm_x = cm_norm.drop('overall', axis = 1)
cm_w = cm_norm['overall']

lm.fit(cm_x,cm_w)
yhat = lm.predict(cm_x)
print(lm.intercept_)
print(lm.coef_)
```

```python
# get the actual and predicted table for cms
avp = pd.DataFrame({'Actual' : cm_w, 'Predicted' : yhat})
df.sort_index(inplace=True)

add_names = avp.join(names_cm['short_name'])
```

```python
# plot the actual vs predicted of top 100 rated centre-midfielders according
# to their potential

add_names[0:200].plot(x = 'short_name', y = ['Actual','Predicted'])
plt.figtext(.5, .9, "Predicted vs Actual 'overall' of Top 200 rated Central Midfielders on FIFA 21", fontsize = 20, ha = 'center')
plt.xlabel('Centre Midfielder Names', fontsize = 17)
plt.xticks(rotation = 45)
plt.ylabel('Overall', fontsize = 17)
plt.show()
```

```python
# get the error rate for linear regression
from sklearn import metrics
from math import sqrt
print('Mean Squared Error:', sqrt(metrics.mean_squared_error(cm_w, yhat)))
print('Accuracy:', 1 - sqrt(metrics.mean_squared_error(cm_w, yhat)))
```

```python
# accuracy for the top 200 centre midfielders
print('Accuracy:', 1 - sqrt(metrics.mean_squared_error(cm_w[:200], yhat[:200])))
```
